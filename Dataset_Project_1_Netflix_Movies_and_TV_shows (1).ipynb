```python
import pandas as pd
d1 = pd.read_csv("Netflix TV Shows and Movies.csv")
movies = d1[d1["type"] == "MOVIE"]
movies_sorted = movies.sort_values("release_year")
movies_1990_2000 = movies_sorted[(movies_sorted["release_year"] >= 1990) & (movies_sorted["release_year"] <= 2000)]
print(movies_1990_2000.iloc[:, :])
```

```python
import pandas as pd
d1 = pd.read_csv("Netflix TV Shows and Movies.csv")
d1 = d1.dropna(subset=["imdb_score"])
high_rated = d1[d1["imdb_score"] > 8.1]
high_rated_sorted = high_rated.sort_values("imdb_score", ascending=False)
print(high_rated_sorted[["title", "type", "release_year", "imdb_score"]])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Netflix TV Shows and Movies.csv")
movies = df[df["type"] == "MOVIE"]
movies = movies.dropna(subset=["imdb_score", "release_year"])
plt.figure(figsize=(10, 6))
plt.scatter(movies["release_year"], movies["imdb_score"])
plt.title("IMDB Ratings vs Release Year (Movies)")
plt.xlabel("Release Year")
plt.ylabel("IMDB Rating")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Netflix TV Shows and Movies.csv")
movies = df[df["type"] == "MOVIE"]
movies = movies.dropna(subset=["imdb_score", "release_year"])
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies, x="release_year", y="imdb_score")
plt.title("IMDB Ratings vs Release Year (Movies)")
plt.xlabel("Release Year")
plt.ylabel("IMDB Rating")
plt.show()
```
